```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'combined_dataset.csv' is the filename
df = pd.read_csv('combined_dataset.csv')

# Step 1: Check for null values
null_values = df.isnull().sum()
print("Null values in each column:\n", null_values)

# Step 2: Decide between CCA and simple imputation
# If there are null values, we need to handle them
if null_values.any():
    # For demonstration, we'll choose simple imputation
    # Step 3: Check the skewness of the distribution for each column
    skewness = df.skew()
    print("\nSkewness of each column:\n", skewness)

    # Step 4: Visualize the distributions
    df.hist(bins=15, figsize=(15, 10), layout=(3, 3))
    plt.tight_layout()
    plt.show()

    # Step 5: Perform imputation based on skewness
    imputed_df = df.copy()
    for column in df.columns:
        if df[column].isnull().any():
            if skewness[column] < -1 or skewness[column] > 1:
                # Highly skewed, use median for imputation
                imputed_df[column].fillna(imputed_df[column].median(), inplace=True)
                print(f"Imputed {column} with median value: {imputed_df[column].median()}")
            else:
                # Less skewed, use mean for imputation
                imputed_df[column].fillna(imputed_df[column].mean(), inplace=True)
                print(f"Imputed {column} with mean value: {imputed_df[column].mean()}")
else:
    df = df.copy()
    print("No null values found, no imputation needed.")

# Step 6: Display the imputed data
print("\nData after imputation:\n", df)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error


# Drop the 'League' column
df.drop('League', axis=1, inplace=True)



# Split the dataset into features (X) and target variable (y)
X = df.drop('Total_Market_Value', axis=1)
y = df['Total_Market_Value']

# Perform feature selection using SelectKBest
selector = SelectKBest(f_regression, k='all')
X_selected = selector.fit_transform(X, y)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# Define the machine learning models
models = [
    ('Linear Regression', LinearRegression()),
    ('Ridge', Ridge()),
    ('Lasso', Lasso()),
    ('Decision Tree', DecisionTreeRegressor()),
    ('Random Forest', RandomForestRegressor())
]

# Use GridSearchCV for hyperparameter tuning
best_models = []
for name, model in models:
    pipeline = Pipeline([
        ('model', model)
    ])
    params = {}
    if name == 'Ridge':
        params = {'model__alpha': [0.1, 1, 10]}
    elif name == 'Lasso':
        params = {'model__alpha': [0.1, 1, 10]}
    elif name == 'Decision Tree':
        params = {'model__max_depth': [None, 5, 10, 20]}
    elif name == 'Random Forest':
        params = {'model__n_estimators': [50, 100, 200]}
    
    grid_search = GridSearchCV(pipeline, param_grid=params, cv=5, scoring='neg_mean_squared_error')
    grid_search.fit(X_train, y_train)
    best_models.append((name, grid_search.best_estimator_))

# Evaluate the models using cross-validation
for name, model in best_models:
    scores = cross_val_score(model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
    rmse_scores = np.sqrt(-scores)
    print(f"{name}: RMSE: {rmse_scores.mean()}")

# Select the best model based on performance
best_model = min(best_models, key=lambda x: cross_val_score(x[1], X_train, y_train, cv=5, scoring='neg_mean_squared_error').mean())
print(f"Best model: {best_model[0]}")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.feature_selection import SelectKBest, f_regression
import pickle

# Split the dataset into features (X) and target variable (y)
X = df.drop('Total_Market_Value', axis=1)
y = df['Total_Market_Value']

# Define the pipeline
pipeline = Pipeline([
    ('feature_selection', SelectKBest(f_regression)),
    ('poly_features', PolynomialFeatures()),  # Add polynomial features
    ('scaler', StandardScaler()),
    ('linear_regression', LinearRegression())
])

# Define the hyperparameters and their values to be tuned
param_grid = {
    'feature_selection__k': ['all', 5, 10],  # Number of features to select
    'poly_features__degree': [1, 2, 3],  # Polynomial degrees to try
    'linear_regression__fit_intercept': [True, False]
}

# Perform grid search to find the best parameters
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='r2')
grid_search.fit(X, y)

# Get the best estimator
best_pipeline = grid_search.best_estimator_

# Generate predictions
y_pred = best_pipeline.predict(X)

# Calculate R-squared score
r2 = r2_score(y, y_pred)
print("R2 score:", r2)

# Calculate mean squared error
mse = mean_squared_error(y, y_pred)
print("Mean Squared Error:", mse)

# Save the best pipeline to a file
with open('football.pkl', 'wb') as f:
    pickle.dump(best_pipeline, f)
```
